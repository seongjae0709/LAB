```python
cols = [
    '사망안전벨트미착용(뒷좌석)',
    '사망안전벨트미착용(운전자)',
    '사망안전벨트미착용(조수석)'
]

# 혹시 문자형이 섞여 있을 경우를 대비
data_2014[cols] = data_2014[cols].apply(pd.to_numeric, errors='coerce').fillna(0)

# 합산 컬럼 생성
data_2014['사망안전벨트미착용'] = data_2014[cols].sum(axis=1)

# 기존 세부 컬럼 삭제
data_2014 = data_2014.drop(columns=cols)
```

### 2015 ~ 2019

```python
data_2015_2019 = pd.read_excel(r'D:\LAB\world ITS\속보자료\2015-2019_교통사고정보_취합분석.xlsx')
data_2015_2019.shape
```

```python
data_2015_2019.columns = data_2015_2019.columns.str.replace('\n', '', regex=False)
```

```python
data_2015_2019.columns
```

```python
cols_not = [
    '사망안전벨트미착용(뒷좌석)',
    '사망안전벨트미착용(운전자)',
    '사망안전벨트미착용(조수석)'
]

cols_yes = [
    '사망안전벨트착용(뒷좌석)',
    '사망안전벨트착용(운전자)',
    '사망안전벨트착용(조수석)'
]

cols_unknown = [
    '사망안전벨트확인불가(뒷좌석)',
    '사망안전벨트확인불가(운전자)',
    '사망안전벨트확인불가(조수석)'
]

all_cols = cols_not + cols_yes + cols_unknown
data_2015_2019[all_cols] = (
    data_2015_2019[all_cols]
    .apply(pd.to_numeric, errors='coerce')
    .fillna(0)
)

data_2015_2019['사망안전벨트미착용'] = data_2015_2019[cols_not].sum(axis=1)
data_2015_2019['사망안전벨트착용']   = data_2015_2019[cols_yes].sum(axis=1)
data_2015_2019['사망안전벨트확인불가'] = data_2015_2019[cols_unknown].sum(axis=1)

data_2015_2019 = data_2015_2019.drop(columns=all_cols)
```

```python
### 원인차운전자연령 -> 원인차운전자연령구분 / 피해차운전자연령 -> 피해차운전자연령구분 으로 각각 열 이름 변경
data_2015_2019 = data_2015_2019.rename(columns={
    '원인차운전자연령': '원인차운전자연령구분',
    '피해차운전자연령': '피해차운전자연령구분'
})
```

```python
# 2011 데이터의 컬럼 이름과 2013 데이터의 컬럼 이름 비교
cols_df_all = df_all.columns
cols_2015_2019 = data_2015_2019.columns


only_in_df2011 = cols_df_all.difference(cols_2015_2019)

only_in_df2012 = cols_2015_2019.difference(cols_df_all)

print("\ndf_all에만 있는 컬럼")
print(*only_in_df2011, sep="\n")

print("\n2015~2019에만 있는 컬럼")
print(*only_in_df2012, sep="\n")
```

```python
### df_all(2011~2014) + data_2015_2019 = df_total
df_total = pd.concat(
    [df_all, data_2015_2019],
    axis=0,
    join='inner',      # 공통 컬럼만 유지
    ignore_index=True  # 인덱스 재정렬
)
```
